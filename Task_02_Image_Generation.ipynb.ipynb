```python
# 1) Install libraries (takes a few minutes)
!pip -q install diffusers==0.29.2 transformers accelerate safetensors torch --extra-index-url https://download.pytorch.org/whl/cu121

# 2) (Optional) Login to Hugging Face if a model requires it
# from huggingface_hub import login
# login()  # paste your HF token when prompted

# 3) Load a Stable Diffusion pipeline (SDXL).
# If VRAM is low or errors occur, switch to "runwayml/stable-diffusion-v1-5" below.
import torch
from diffusers import AutoPipelineForText2Image

model_id = "stabilityai/stable-diffusion-xl-base-1.0"   # fallback: "runwayml/stable-diffusion-v1-5"
pipe = AutoPipelineForText2Image.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    variant="fp16"
).to("cuda")

# 4) Generate one image from a prompt
prompt = "ultra-detailed cinematic photo of a cyberpunk street at night with neon signs, rain reflections, 35mm depth of field"
negative_prompt = "low quality, blurry, watermark, extra fingers, text, logo"
generator = torch.Generator(device="cuda").manual_seed(42)

image = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=28,
    guidance_scale=6.5,
    width=1024, height=1024,   # use 512x512 if you get CUDA OOM
    generator=generator
).images[0]

# 5) Save the output
import os
os.makedirs("/content/outputs", exist_ok=True)
save_path = "/content/outputs/cyberpunk_sdxl_seed42.png"
image.save(save_path)
print("Saved to:", save_path)

```

```python
prompts = [
  "studio portrait of an astronaut in a garden, soft rim light, kodak portra",
  "isometric cozy room, pastel colors, plants and a cat, artstation trending",
  "ancient temple ruins in jungle, golden hour, volumetric light, 8k"
]
neg = "blurry, lowres, watermark, text, logo, deformed"

for i, p in enumerate(prompts, 1):
    img = pipe(p, negative_prompt=neg, num_inference_steps=24, guidance_scale=7).images[0]
    out = f"/content/outputs/img_{i}.png"
    img.save(out)
    print("Saved:", out, " | Prompt:", p)
```
